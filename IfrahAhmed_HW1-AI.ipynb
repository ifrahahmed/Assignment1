```python
##Load the ZHVI data for each ZIP code, selecting only Philadelphia ZIP codes.
```

```python
philly_df = zillow_df.loc[zillow_df['City'] =='Philadelphia']
```

```python
philly_df
```

```python
philly_df['RegionName'].unique()
```

```python
x = ['SizeRank', 'RegionID', "RegionType", "StateName","State","Metro","CountyName","City"] 
philly_final = philly_df.drop(x, axis=1).reset_index(drop=True)
```

```python
philly_final
```

```python
ZHVI_philly = pd.melt(
    philly_final, 
    id_vars=["RegionName"], 
    value_name="ZHVI", 
    var_name="Date"
)
```

```python
##Calculate the annual average ZHVI for each ZIP code in Philadelphia and each year.
```

```python
ZHVI_philly
```

```python
ZHVI_philly["Date"] = pd.to_datetime(ZHVI_philly["Date"])
```

```python
date_strings = ZHVI_philly["Date"].dt.strftime("%Y")
```

```python
ZHVI_philly["Date"] = date_strings
```

```python
ZHVI_philly_mean = ZHVI_philly.groupby(['RegionName','Date']).mean()
```

```python
ZHVI_philly_mean = ZHVI_philly_mean.sort_values(by = ['Date','ZHVI']).reset_index()
```

```python
ZHVI_philly_mean
```

```python
##Identify the ZIP code with the maximum value for each year
```

```python
ZHVI_max = ZHVI_philly_mean.groupby('Date').max().reset_index()
```
